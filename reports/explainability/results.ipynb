```python
import seaborn as sns

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df['group'] = np.where((df['IT_ED'] == 0) & (df['IT_WORK'] == 0), 'no IT experience', 'IT experience')
grouped = df.groupby('group')[rating_cols].mean()
overall = df[rating_cols].mean().to_frame().T
overall.index = ['all']
grouped_df = pd.concat([grouped, overall])

split_columns = grouped_df.columns.str.extract(r'(?P<Algorithm>.*?): (?P<Dataset>.*?) - (?P<Metric>.*)')
split_columns.index = grouped_df.columns

algorithm_map = {
    'Collaborative Filtering BPR': 'CF',
    'Content-based Filtering': 'CBF',
    'Reinforcement Learning': 'RL'
}
split_columns['Algorithm'] = split_columns['Algorithm'].replace(algorithm_map)
grouped_df.columns = pd.MultiIndex.from_frame(split_columns)

def plot_group_histograms_one_fig_per_dataset(group):
    metric_map = {
        'Satysfakcja': 'Satisfaction',
        'Proste': 'Simple',
        'Elastyczne': 'Adaptable',
        'Ugruntowane': 'Grounded',
        'Rozszerzalne': 'Extendable',
        'Oparte na źródłach': 'Source-based'
    }
    dataset_map = {
        'Oferty kupna produktów': 'amazonsales',
        'Filmy': 'movielens',
        'Posty': 'postrec'
    }
    ordered_metrics_pl = [
        'Satysfakcja',
        'Proste',
        'Elastyczne',
        'Ugruntowane',
        'Rozszerzalne',
        'Oparte na źródłach'
    ]

    palette = sns.color_palette('Set2')

    for dataset in group.columns.levels[1]:
        dataset_data = group.xs(dataset, level='Dataset', axis=1)
        if dataset_data.empty:
            continue

        metrics = ordered_metrics_pl
        n_metrics = len(metrics)
        n_cols = 3
        n_rows = (n_metrics + n_cols - 1) // n_cols

        fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 4, n_rows * 4), squeeze=False)

        for idx, metric_pl in enumerate(metrics):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col]

            try:
                metric_data = dataset_data.xs(metric_pl, level='Metric', axis=1)
            except KeyError:
                ax.axis('off')
                continue

            if metric_data.empty:
                ax.axis('off')
                continue

            alg_labels = metric_data.columns.get_level_values('Algorithm')
            width = 0.8 / len(metric_data.columns)
            x = np.arange(len(metric_data.index))

            for i, col_name in enumerate(metric_data.columns):
                alg = alg_labels[i]
                heights = metric_data.iloc[:, i]
                bar_positions = x - 0.4 + i * width + width / 2
                bars = ax.bar(bar_positions, heights, width=width, label=alg, color=palette[i % len(palette)])

                for j, bar in enumerate(bars):
                    ax.text(
                        bar.get_x() + bar.get_width() / 2,
                        bar.get_height() + 0.01,
                        alg,
                        ha='center', va='bottom', fontsize=10, color='black'
                    )

            ax.set_xticks(x)
            ax.set_xticklabels(metric_data.index)
            ax.set_ylim(0, 5)
            ax.yaxis.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

            metric_en = metric_map.get(metric_pl, metric_pl)
            ax.set_title(metric_en)

        for idx in range(n_metrics, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            axes[row, col].axis('off')

        dataset_en = dataset_map.get(dataset, dataset)
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.savefig(f'fig/{dataset_en}.png')
        plt.show()


plot_group_histograms_one_fig_per_dataset(grouped_df)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df['group'] = np.where((df['IT_ED'] == 0) & (df['IT_WORK'] == 0), 'no IT experience', 'IT experience')
grouped = df.groupby('group')[rating_cols].mean()
overall = df[rating_cols].mean().to_frame().T
overall.index = ['all']
grouped_df = pd.concat([grouped, overall])

split_columns = grouped_df.columns.str.extract(r'(?P<Algorithm>.*?): (?P<Dataset>.*?) - (?P<Metric>.*)')
split_columns.index = grouped_df.columns

algorithm_map = {
    'Collaborative Filtering BPR': 'CF',
    'Content-based Filtering': 'CBF',
    'Reinforcement Learning': 'RL'
}
split_columns['Algorithm'] = split_columns['Algorithm'].replace(algorithm_map)
grouped_df.columns = pd.MultiIndex.from_frame(split_columns)

grouped_avg = grouped_df.groupby(level=['Algorithm', 'Metric'], axis=1).mean()

def plot_histograms_by_algorithm(group_avg):
    metric_map = {
        'Satysfakcja': 'Satisfaction',
        'Proste': 'Simple',
        'Elastyczne': 'Adaptable',
        'Ugruntowane': 'Grounded',
        'Rozszerzalne': 'Extendable',
        'Oparte na źródłach': 'Source-based'
    }
    ordered_metrics_pl = [
        'Satysfakcja',
        'Proste',
        'Elastyczne',
        'Ugruntowane',
        'Rozszerzalne',
        'Oparte na źródłach'
    ]

    palette = sns.color_palette('Set2')

    n_metrics = len(ordered_metrics_pl)
    n_cols = 3
    n_rows = (n_metrics + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 4, n_rows * 4), squeeze=False)

    for idx, metric_pl in enumerate(ordered_metrics_pl):
        row = idx // n_cols
        col = idx % n_cols
        ax = axes[row, col]

        metric_data = group_avg.xs(metric_pl, level='Metric', axis=1)

        if metric_data.empty:
            ax.axis('off')
            continue

        alg_labels = metric_data.columns
        width = 0.8 / len(alg_labels)
        x = np.arange(len(metric_data.index))

        for i, alg in enumerate(alg_labels):
            heights = metric_data.iloc[:, i]
            bar_positions = x - 0.4 + i * width + width / 2
            bars = ax.bar(bar_positions, heights, width=width, label=alg, color=palette[i % len(palette)])

            for j, bar in enumerate(bars):
                ax.text(
                    bar.get_x() + bar.get_width() / 2,
                    bar.get_height() + 0.01,
                    alg,
                    ha='center', va='bottom', fontsize=10, color='black'
                )

        ax.set_xticks(x)
        ax.set_xticklabels(metric_data.index)
        ax.set_ylim(0, 5)
        ax.yaxis.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

        metric_en = metric_map.get(metric_pl, metric_pl)
        ax.set_title(metric_en)

    for idx in range(n_metrics, n_rows * n_cols):
        row = idx // n_cols
        col = idx % n_cols
        axes[row, col].axis('off')

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.savefig(f'fig/all.png')
    plt.show()

plot_histograms_by_algorithm(grouped_avg)
```

```python
import re

metric_cols = [col for col in df.columns if '-' in col]

result = {}

for col in metric_cols:
    match = re.match(r'(.+?):.+?- (.+)', col)
    if not match:
        continue
    algorithm = match.group(1).strip()
    metric = match.group(2).strip()

    value_counts = df[col].value_counts(normalize=True) * 100
    value_counts = value_counts.sort_index().round(2)

    if algorithm not in result:
        result[algorithm] = {}
    result[algorithm][metric] = value_counts.to_dict()

for alg, metrics in result.items():
    print(f'\n### Algorytm: {alg}')
    for metric, scores in metrics.items():
        print(f'\n  - Metryka: {metric}')
        for score, percent in scores.items():
            print(f'    Ocena {score}: {percent}%')
```

```python
import pandas as pd


def classify_it_experience(edu, work):
    if edu == 'Tak' and work == 'Tak':
        return 'Advanced IT Background'
    
    elif edu == 'Częściowo' and work == 'Tak':
        return 'Strong Practical Background'
    elif edu == 'Tak' and work == 'Częściowo':
        return 'Strong Academic Background'
    
    elif edu == 'Częściowo' and work == 'Częściowo':
        return 'Intermediate IT Background'
    
    elif edu == 'Tak' and work == 'Nie':
        return 'Academic-Only IT Background'
    elif edu == 'Nie' and work == 'Tak':
        return 'Work-Only IT Background'
    
    elif edu == 'Częściowo' and work == 'Nie':
        return 'Basic Academic Exposure'
    elif edu == 'Nie' and work == 'Częściowo':
        return 'Basic Practical Exposure'
    
    elif edu == 'Nie' and work == 'Nie':
        return 'No IT Background'
    
    else:
        return 'Unclassified'


df['IT_EXPERIENCE_GROUP'] = df.apply(lambda row: classify_it_experience(row['IT education'], row['IT work']), axis=1)

df['IT_EXPERIENCE_GROUP'].value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Arc


group_counts = df['IT_EXPERIENCE_GROUP'].value_counts()

custom_colors = {
    'No IT Background': '#FF4C4C',
    'Basic Academic Exposure': "#4C6FAF",
    'Strong Academic Background': "#4CAFAF",
    'Basic Practical Exposure': "#4CAF70",
    'Advanced IT Background': "#8BAF4C",
}

colors = [custom_colors.get(label, '#CCCCCC') for label in group_counts.index]

fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(
    group_counts,
    labels=group_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    labeldistance=1.04
)

plt.axis('equal')
plt.tight_layout()

no_it_index = list(group_counts.index).index('No IT Background')
wedge_no_it = wedges[no_it_index]

theta1_no_it, theta2_no_it = wedge_no_it.theta1, wedge_no_it.theta2
center = wedge_no_it.center
r = wedge_no_it.r
outer_r = r + 0.01

arc_no_it = Arc(center, 2*outer_r, 2*outer_r,
                angle=0, theta1=theta1_no_it, theta2=theta2_no_it,
                linewidth=3, color='red')

arc_it_1 = Arc(center, 2*outer_r, 2*outer_r,
               angle=0, theta1=theta2_no_it, theta2=theta1_no_it + 360,
               linewidth=3, color='green')

ax.add_patch(arc_no_it)
ax.add_patch(arc_it_1)

theta_mid_no_it = (theta1_no_it + theta2_no_it) / 2
x_no_it = center[0] + (outer_r - 0.1) * np.cos(np.deg2rad(theta_mid_no_it))
y_no_it = center[1] + (outer_r - 0.1) * np.sin(np.deg2rad(theta_mid_no_it))
ax.text(x_no_it, y_no_it, "No IT experience", color='red', fontsize=12, fontweight='bold',
        ha='center', va='center', bbox=dict(facecolor='white', edgecolor='none', alpha=0.7, pad=2))

theta_mid_it = (theta2_no_it + (theta1_no_it + 360)) / 2
x_it = center[0] + (outer_r - 0.1) * np.cos(np.deg2rad(theta_mid_it))
y_it = center[1] + (outer_r - 0.1) * np.sin(np.deg2rad(theta_mid_it))
ax.text(x_it, y_it, "IT experience", color='green', fontsize=12, fontweight='bold',
        ha='center', va='center', bbox=dict(facecolor='white', edgecolor='none', alpha=0.7, pad=2))

plt.savefig(f'fig/respondents.png')
plt.show()
```
