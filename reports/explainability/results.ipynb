```python
import seaborn as sns

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df['group'] = np.where((df['IT_ED'] == 0) & (df['IT_WORK'] == 0), 'no IT experience', 'IT experience')
grouped = df.groupby('group')[rating_cols].mean()
overall = df[rating_cols].mean().to_frame().T
overall.index = ['all']
grouped_df = pd.concat([grouped, overall])

split_columns = grouped_df.columns.str.extract(r'(?P<Algorithm>.*?): (?P<Dataset>.*?) - (?P<Metric>.*)')
split_columns.index = grouped_df.columns

algorithm_map = {
    'Collaborative Filtering BPR': 'CF',
    'Content-based Filtering': 'CBF',
    'Reinforcement Learning': 'RL'
}
split_columns['Algorithm'] = split_columns['Algorithm'].replace(algorithm_map)
grouped_df.columns = pd.MultiIndex.from_frame(split_columns)

def plot_group_histograms_one_fig_per_dataset(group):
    metric_map = {
        'Satysfakcja': 'Satisfaction',
        'Proste': 'Simple',
        'Elastyczne': 'Flexible',
        'Ugruntowane': 'Grounded',
        'Rozszerzalne': 'Extendable',
        'Oparte na źródłach': 'Source-based'
    }
    dataset_map = {
        'Oferty kupna produktów': 'amazonsales',
        'Filmy': 'movielens',
        'Posty': 'postrec'
    }
    ordered_metrics_pl = [
        'Satysfakcja',
        'Proste',
        'Elastyczne',
        'Ugruntowane',
        'Rozszerzalne',
        'Oparte na źródłach'
    ]

    palette = sns.color_palette('Set2')

    for dataset in group.columns.levels[1]:
        dataset_data = group.xs(dataset, level='Dataset', axis=1)
        if dataset_data.empty:
            continue

        metrics = ordered_metrics_pl
        n_metrics = len(metrics)
        n_cols = 3
        n_rows = (n_metrics + n_cols - 1) // n_cols

        fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 4, n_rows * 4), squeeze=False)

        for idx, metric_pl in enumerate(metrics):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col]

            try:
                metric_data = dataset_data.xs(metric_pl, level='Metric', axis=1)
            except KeyError:
                ax.axis('off')
                continue

            if metric_data.empty:
                ax.axis('off')
                continue

            alg_labels = metric_data.columns.get_level_values('Algorithm')
            width = 0.8 / len(metric_data.columns)
            x = np.arange(len(metric_data.index))

            for i, col_name in enumerate(metric_data.columns):
                alg = alg_labels[i]
                heights = metric_data.iloc[:, i]
                bar_positions = x - 0.4 + i * width + width / 2
                bars = ax.bar(bar_positions, heights, width=width, label=alg, color=palette[i % len(palette)])

                for j, bar in enumerate(bars):
                    ax.text(
                        bar.get_x() + bar.get_width() / 2,
                        bar.get_height() + 0.01,
                        alg,
                        ha='center', va='bottom', fontsize=10, color='black'
                    )

            ax.set_xticks(x)
            ax.set_xticklabels(metric_data.index)
            ax.set_ylim(0, 5)
            ax.yaxis.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

            metric_en = metric_map.get(metric_pl, metric_pl)
            ax.set_title(metric_en)

        for idx in range(n_metrics, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            axes[row, col].axis('off')

        dataset_en = dataset_map.get(dataset, dataset)
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        plt.savefig(f'fig/{dataset_en}.png')
        plt.show()


plot_group_histograms_one_fig_per_dataset(grouped_df)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df['group'] = np.where((df['IT_ED'] == 0) & (df['IT_WORK'] == 0), 'no IT experience', 'IT experience')
grouped = df.groupby('group')[rating_cols].mean()
overall = df[rating_cols].mean().to_frame().T
overall.index = ['all']
grouped_df = pd.concat([grouped, overall])

split_columns = grouped_df.columns.str.extract(r'(?P<Algorithm>.*?): (?P<Dataset>.*?) - (?P<Metric>.*)')
split_columns.index = grouped_df.columns

algorithm_map = {
    'Collaborative Filtering BPR': 'CF',
    'Content-based Filtering': 'CBF',
    'Reinforcement Learning': 'RL'
}
split_columns['Algorithm'] = split_columns['Algorithm'].replace(algorithm_map)
grouped_df.columns = pd.MultiIndex.from_frame(split_columns)

grouped_avg = grouped_df.groupby(level=['Algorithm', 'Metric'], axis=1).mean()

def plot_histograms_by_algorithm(group_avg):
    metric_map = {
        'Satysfakcja': 'Satisfaction',
        'Proste': 'Simple',
        'Elastyczne': 'Flexible',
        'Ugruntowane': 'Grounded',
        'Rozszerzalne': 'Extendable',
        'Oparte na źródłach': 'Source-based'
    }
    ordered_metrics_pl = [
        'Satysfakcja',
        'Proste',
        'Elastyczne',
        'Ugruntowane',
        'Rozszerzalne',
        'Oparte na źródłach'
    ]

    palette = sns.color_palette('Set2')

    n_metrics = len(ordered_metrics_pl)
    n_cols = 3
    n_rows = (n_metrics + n_cols - 1) // n_cols

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 4, n_rows * 4), squeeze=False)

    for idx, metric_pl in enumerate(ordered_metrics_pl):
        row = idx // n_cols
        col = idx % n_cols
        ax = axes[row, col]

        metric_data = group_avg.xs(metric_pl, level='Metric', axis=1)

        if metric_data.empty:
            ax.axis('off')
            continue

        alg_labels = metric_data.columns
        width = 0.8 / len(alg_labels)
        x = np.arange(len(metric_data.index))

        for i, alg in enumerate(alg_labels):
            heights = metric_data.iloc[:, i]
            bar_positions = x - 0.4 + i * width + width / 2
            bars = ax.bar(bar_positions, heights, width=width, label=alg, color=palette[i % len(palette)])

            for j, bar in enumerate(bars):
                ax.text(
                    bar.get_x() + bar.get_width() / 2,
                    bar.get_height() + 0.01,
                    alg,
                    ha='center', va='bottom', fontsize=10, color='black'
                )

        ax.set_xticks(x)
        ax.set_xticklabels(metric_data.index)
        ax.set_ylim(0, 5)
        ax.yaxis.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

        metric_en = metric_map.get(metric_pl, metric_pl)
        ax.set_title(metric_en)

    for idx in range(n_metrics, n_rows * n_cols):
        row = idx // n_cols
        col = idx % n_cols
        axes[row, col].axis('off')

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.savefig(f'fig/all.png')
    plt.show()

plot_histograms_by_algorithm(grouped_avg)
```

```python
def plot_group_histograms_one_fig_per_dataset(group):
    metric_map = {
        'Satysfakcja': 'Satisfaction',
        'Proste': 'Simple',
        'Elastyczne': 'Flexible',
        'Ugruntowane': 'Grounded',
        'Rozszerzalne': 'Extendable',
        'Oparte na źródłach': 'Source-based'
    }
    dataset_map = {
        'Oferty kupna produktów': 'amazonsales',
        'Filmy': 'movielens',
        'Posty': 'postrec'
    }
    ordered_metrics_pl = [
        'Satysfakcja',
        'Proste',
        'Elastyczne',
        'Ugruntowane',
        'Rozszerzalne',
        'Oparte na źródłach'
    ]
    line_colors = {
        'Oferty kupna produktów': 'blue',
        'Filmy': 'green',
        'Posty': '#CC5500'
    }

    palette = sns.color_palette('Set2')

    # Przygotuj słownik ze średnimi wartościami dla wszystkich datasetów
    dataset_avg = {}
    for dataset in group.columns.levels[1]:
        dataset_data = group.xs(dataset, level='Dataset', axis=1)
        if dataset_data.empty:
            continue
        dataset_avg[dataset] = dataset_data

    for dataset in group.columns.levels[1]:
        dataset_data = group.xs(dataset, level='Dataset', axis=1)
        if dataset_data.empty:
            continue

        metrics = ordered_metrics_pl
        n_metrics = len(metrics)
        n_cols = 3
        n_rows = (n_metrics + n_cols - 1) // n_cols

        fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols * 4, n_rows * 4), squeeze=False)

        for idx, metric_pl in enumerate(metrics):
            row = idx // n_cols
            col = idx % n_cols
            ax = axes[row, col]

            try:
                metric_data = dataset_data.xs(metric_pl, level='Metric', axis=1)
            except KeyError:
                ax.axis('off')
                continue

            if metric_data.empty:
                ax.axis('off')
                continue

            alg_labels = metric_data.columns.get_level_values('Algorithm')
            width = 0.8 / len(metric_data.columns)
            x = np.arange(len(metric_data.index))

            # Słupki
            for i, col_name in enumerate(metric_data.columns):
                alg = alg_labels[i]
                heights = metric_data.iloc[:, i]
                bar_positions = x - 0.4 + i * width + width / 2
                bars = ax.bar(bar_positions, heights, width=width, color=palette[i % len(palette)])

                for j, bar in enumerate(bars):
                    ax.text(
                        bar.get_x() + bar.get_width() / 2,
                        bar.get_height() + 0.01,
                        alg,
                        ha='center', va='bottom', fontsize=10, color='black', fontweight='bold'
                    )

                for other_dataset, other_data in dataset_avg.items():
                    if other_dataset == dataset:
                        continue
                    try:
                        other_metric_data = other_data.xs(metric_pl, level='Metric', axis=1)
                    except KeyError:
                        continue
                    if other_metric_data.empty or col_name not in other_metric_data.columns:
                        continue

                    y_val = other_metric_data[col_name].mean()
                    color = line_colors.get(other_dataset, 'gray')

                    for pos in bar_positions:
                        ax.hlines(y_val,
                                xmin=pos - width / 2,
                                xmax=pos + width / 2,
                                linestyle='--', color=color, alpha=0.7,
                                label=dataset_map.get(other_dataset, other_dataset) if idx == 0 and i == 0 and pos == bar_positions[0] else None)

            ax.set_xticks(x)
            ax.set_xticklabels(metric_data.index)
            ax.set_ylim(0, 5)
            ax.yaxis.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

            metric_en = metric_map.get(metric_pl, metric_pl)
            ax.set_title(metric_en)

        # Dodaj legendę (bez duplikatów)
        handles, labels = axes[0, 0].get_legend_handles_labels()
        by_label = dict(zip(labels, handles))
        fig.legend(by_label.values(), by_label.keys(), loc='upper right', bbox_to_anchor=(1, 0.95))

        for idx in range(n_metrics, n_rows * n_cols):
            row = idx // n_cols
            col = idx % n_cols
            axes[row, col].axis('off')

        dataset_en = dataset_map.get(dataset, dataset)
        plt.tight_layout(rect=[0, 0, 1, 0.95])
        # plt.savefig(f'fig/{dataset_en}.png')
        plt.show()
plot_group_histograms_one_fig_per_dataset(grouped_df)
```
